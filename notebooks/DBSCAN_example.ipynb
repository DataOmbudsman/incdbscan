```python
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs, make_circles

from src.dbscan import DBSCAN
```

# Clustering Gaussian blobs

```python
X_blobs, blob_labels = make_blobs(n_samples=100, centers=5, n_features=2, cluster_std=0.3, random_state=123)
```

## Visualize true labels

```python
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=blob_labels);
```

## Cluster and visualize clustering results

```python
clustering_blob = DBSCAN(eps=1, min_pts=5)
clustering_blob.fit(X_blobs)
```

```python
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=clustering_blob.labels);
```

# Clustering circles

```python
X_circles, circle_labels = make_circles(n_samples=200, noise=0.05, factor=0.5, random_state=123)

# add noise
X_circles[-1, :] = [0, 0]
circle_labels[-1] = -1
```

## Visualize true labels

```python
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=circle_labels)
```

## Cluster and visualize clustering results

```python
clustering_circles = DBSCAN(eps=0.3, min_pts=5)
clustering_circles.fit(X_circles)
```

```python
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=clustering_circles.labels);
```
